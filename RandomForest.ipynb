```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
train = pd.read_csv('train.csv') 
train.head()
```

```python
train.shape
```

```python
train.isnull().sum()
```

```python
na_df = pd.concat([train.isnull().sum()],axis = 1,keys=['Count'])
na_df
```

```python
na_df[na_df['Count']>0]
```

```python
train['Age'] = train['Age'].fillna(train['Age'].mean())
```

```python
train['Cabin'].unique() #.mode()[0]
```

```python
train['Cabin'] = train['Cabin'].fillna(train['Cabin'].mode()[0])
```

```python
train.head()
```

```python
train['Embarked'] = train['Embarked'].fillna(train['Embarked'].mode()[0])
```

```python
train.isnull().sum()
```

```python
train2 = train.drop(['Name','Ticket'],axis = 1)
train2.head()
```

```python
train3 =  pd.concat([train2,pd.get_dummies(train2['Sex'])],axis=1)
train3.head()
```

```python
train3 = train3.drop('Sex',axis = 1)
train3.head()
```

```python
train4= pd.concat([train3,pd.get_dummies(train3['Cabin'])],axis = 1)
train4.head()
```

```python
train4 = train4.drop('Cabin',axis = 1)
train4.head()
```

```python
train4.Embarked.unique()
```

```python
train4['Embarked'] = train4['Embarked'].map({'S':0,'C':1,'Q':2})
```

```python
x = train4.drop('Survived',axis = 1)
y = train4.Survived
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
mdl = RandomForestClassifier()
```

```python
mdl.fit(x_train,y_train)
```

```python
y_pred = mdl.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```
